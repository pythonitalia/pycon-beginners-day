### Task 2.b Use a function to establish a similarity degree

Theory related concepts:
- complex mathematical operators and operators precedence
- math module
- functions
  - input/output of a function
  - parameters

````python
# Code Here:
# - function to calculate Euclidean Distance
# - data normalisation

# ++ Extract only numerical columns for each row in the 

peppo =     [ 54, 55, 32 ]
bulbasaur = [ 45, 49, 49 ]
def euclidean_distance(p1, p2): # def is a keyword to DEFINE a function while p1 and p2 are the parameters, i.e. the input data of the function
  dim = len(p1)
  distance = 0
  for d in range(dim):
    diff = p1[d] - p2[d]
    # print(f"Computing difference -> {p1[d]} - {p2[d]} = {diff}")
    sq = diff ** 2
    # print(f"Square of the difference -> {sq}")
    distance = distance + sq

  # print(f"sum of the squares -> {distance}")

  distance = distance ** (1/2)

  # print(f"square root of the sum of the squares -> {distance}")
    
  return distance

dist = euclidean_distance(peppo, bulbasaur)

print(dist)
````

````python
# Code here: (following up from previous tasks)
# - Get user input, and define an imaginary pokemon_ from the 6 given input values
# - Add the imaginary pokemon to the dataset, and run normalisation
# - Extract the imaginary pokemon from the dataset (last row)
# - calculate the distance (similarity) of the imaginary pokemon from the other pokemon in the dataset
# - sorting of similarities / distances (via lambda function)
# - extract the top K results

l = [40, 30, 60, 80, 25]
print(l)
l.sort()
print(l)

l = [ ["Giuseppe", 2.8] , ["Mark", 1.5], ["Susan", 0.9], ["Barbara", 1] ]

print(l)
l.sort()
print(l)

def peppo(item):
  return item[1]

l.sort(key=peppo)

print(l)
````

````python
pepposaur = [ hp, attack, defense, sp_attack, sp_defense, speed ]

def get_k_neighbors(k, dataset, input_row):
  distances = []

  for index, pokemon in enumerate(dataset):
    d = euclidean_distance(input_row, pokemon)
    obj = [index, d]

    distances.append(obj)

    # print(obj)

  def distance_attribute(item):
    return item[1]

  distances.sort(key=distance_attribute)

  return distances[0:k]

nearest_pokemon = get_k_neighbors(3, numerical_data, pepposaur)

print(nearest_pokemon)


````
